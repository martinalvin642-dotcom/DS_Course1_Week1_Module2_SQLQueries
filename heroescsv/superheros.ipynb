```python
# ============================================================
# Distribution by publisher
# ============================================================

# Clean & count
df['Publisher'] = df['Publisher'].str.strip()
publisher_counts = df['Publisher'].value_counts().dropna().head(10)
publisher_counts
print(publisher_counts)

# Plot
publisher_counts[::-1].plot(kind='barh', figsize=(9,5), color='red',
                             title='Superheroes by Publisher')
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```

```python
# ============================================================
#  Height vs superpowers, by gender
# ============================================================

# Count powers per hero
power_counts = (df1 == True).sum(axis=0).reset_index()
power_counts.columns = ['name', 'power_count']

# Join with heroes
merged = df.merge(power_counts, on='name')

# Clean height (-99 = missing) and keep Male/Female only
merged['Height'] = merged['Height'].replace(-99, np.nan)
merged = merged.dropna(subset=['Height'])
merged = merged[merged['Gender'].isin(['Male', 'Female'])]

# Plot
fig, ax = plt.subplots(figsize=(9, 5))
colors = {'Male': 'steelblue', 'Female': 'coral'}

for gender, group in merged.groupby('Gender'):
    ax.scatter(group['Height'], group['power_count'],
               label=gender, alpha=0.5, color=colors[gender])
    z = np.polyfit(group['Height'], group['power_count'], 1)
    x = np.linspace(group['Height'].min(), group['Height'].max(), 100)
    ax.plot(x, np.poly1d(z)(x), color=colors[gender], linestyle='--')

ax.set(xlabel='Height (cm)', ylabel='Power Count',
       title='Height vs Superpowers by Gender')
ax.legend()
plt.tight_layout()
plt.show()

# Correlation
for gender, group in merged.groupby('Gender'):
    print(f"{gender}: r = {group['Height'].corr(group['power_count']):.3f}")
```

```python
# ============================================================
#  Top 5 powers: Marvel vs DC
# ============================================================

# Hero name sets per publisher
marvel = set(df[df['Publisher'] == 'Marvel Comics']['name'])
dc     = set(df[df['Publisher'] == 'DC Comics']['name'])

# Count powers (powers_df columns = hero names)
marvel_top5 = (df1[[c for c in df1.columns if c in marvel]] == True).sum(axis=1).nlargest(5)
dc_top5     = (df1[[c for c in df1.columns if c in dc]]     == True).sum(axis=1).nlargest(5)

# Plot side by side
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
for ax, data, title, color in zip(axes,
    [marvel_top5, dc_top5],
    ['Marvel — Top 5 Powers', 'DC — Top 5 Powers'],
    ['#E23636', '#0476F5']):

    data[::-1].plot(kind='barh', ax=ax, color=color, title=title)
    ax.set_xlabel('Hero Count')

plt.tight_layout()
plt.show()
```

```python
# ============================================================
# Which alignment has the most powers on average
# ============================================================

result = (merged[merged['Alignment'].isin(['good', 'bad', 'neutral'])]
          .groupby('Alignment')['power_count']
          .mean()
          .sort_values(ascending=False))

print(result)

result.plot(kind='bar', figsize=(7, 4), color=["#2ecc71","#e13926","#75b2b7"],
            title='Avg Powers by Alignment', rot=0)
plt.ylabel('Avg Power Count')
plt.tight_layout()
plt.show()
```
