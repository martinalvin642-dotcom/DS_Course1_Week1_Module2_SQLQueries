### **Introduction to Data Visualization**

Data Visualization is the process of representing data using charts, plots, graphs in a way that is easily interpretable by humans.
Helps in understanding the data, finding patterns, detecting outliers, decision-making etc

Examples: Marketing and Sales company using visuals check which age-groups buy which products, A Basketball Team using plots to compare shot attempts and success rate.

<b>Common Visualization Libraries</b><br>
i. Matplotlib <br>
Foundational visualization library in Python with many types of visuals/charts.<br>

ii. Seaborn<br>
Builds on top of Matplotlib. Makes statistical plots easier and more intuitive.<br>

iii. Plotly<br>
Python library for creating visualizations and dashboards.

Pandas is mainly for data handling but it can create visualizations.

### **Visualizing Different Variable Types**

Before one chooses a visualization, it's important that they understand the type of variable to plot.
The two broad categories of data types are:
1. Numerical
    i. Discrete (countable, whole numbers): age, no of people/students, salary
    ii. Continous (Measurable, within range, decimals): height, weight, temperature

2. Categorical
    i. Nominal (No order): blood group, colour, gender
    ii. Ordinal (Order, not measurable): education level, ratings

<b>Categorical Variables</b>

Bar Chart, Pie Charts

<b>Numerical Variables</b>

Histogram, Box Plots, Scatter Plots (two continous), Heat Map (multiple continous), line graphs (trends)

### **Common Plots**
1. Bar Chart
Used when you are comparing categorical variables.
Example: Which fruit is liked most.

import matplotlib.pyplot as plt
plt.bar()

2. Histograms
Used to show distribution of one continous numeric variables. It helps show where most values are concentrated using bins. The more bins the more the detail, fewer bins show broader grouping.
Example: How are age-groups distributed

import matplotlib.pyplot as plt
plt.hist()

3. Scatter Plot
Used when comparing two continous variables.
Example: Is retirement age related to life expectancy

import matplotlib.pyplot as plt
plt.scatter()

4. Line Graph
Used to show change over time.
Example: Stock price change over a period of time

import matplotlib.pyplot as plt
plt.plot()

5. Box Plot
Used to summarize the spread of one numeric variable and show possible outliers
Example: Unusual age values

import matplotlib.pyplot as plt
plt.boxplot()

```python
# Bar Graph
import matplotlib.pyplot as plt

fruits = ['Apple', 'Banana', 'Orange', 'Grape']
counts = [25, 20, 15, 10]

plt.figure(figsize=(6, 4))
plt.bar(fruits, counts)
plt.xlabel('Fruit')
plt.ylabel('Count')
plt.title('Favorite Fruits')
plt.show()
```

```python
# Histogram
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(100)
x = 5 * np.random.randn(200) + 60

plt.hist(x, bins=10, edgecolor='black')
plt.xlabel('Retirement Age')
plt.ylabel('Frequency')
plt.title('Histogram of Retirement Age')
plt.show()
```
