```python
import matplotlib.pyplot as plt
import numpy as np

ypoints1 = np.array([3,8,2,7])
ypoints2 = np.array([1,4,7,2])
plt.plot(ypoints1,label='line1', linestyle = 'dashed',  marker='o',markerfacecolor='red',markersize=10)
plt.plot(ypoints2,label='line2', linestyle = 'dotted',marker='s',markerfacecolor='blue',markersize=10)
font1= {'family':'arial','color':'green','size':12}
plt.title("DATA", fontweight= 'bold',color = 'red',)
plt.ylabel("data y-axis",fontdict= font1)
plt.legend()


plt.show()
```

scatter plot

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([20,12,35,67,89,45]) #age
y = np.array([45,69,40,70,80,40]) #weight

plt.scatter(x,y)
plt.show()

```

histogram


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```
