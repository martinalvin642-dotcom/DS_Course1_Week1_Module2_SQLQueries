```python
# how to establish a conecction
#1. import sqlite3
#2. import pandas(to display the item in a tabular format.)
#3. call the connection path
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("bookstore.db")

print(conn)
```

```python
# using cursor
cur = conn.cursor()
cur.execute("SELECT * FROM books")
results = cur.fetchall()# it will give us a tuple of all the books
print(results)
```

```python
# exploring database: listing all the tables in  our database
pd.read_sql("SELECT name FROM  sqlite_master WHERE type = 'table';",
            conn
)
```

```python
#  geting authors
pd.read_sql("SELECT * FROM authors",
            conn)
```

```python
# renaming using AS(aliases): create temporary display names
pd.read_sql("""
        SELECT
            title AS book_title,
            price AS price_usd
        FROM books;




""",conn)
```

```python
# select author and from authors table give name an alias of author_name
pd.read_sql("""
        SELECT
            name AS author_name
        FROM authors;


""",conn)
```

```python
# AND & OR (combining conditions)
pd.read_sql("""
   SELECT title, genre, price
   FROM books
   WHERE genre = 'Literary Fiction'        
   AND price < 12.0;
""",conn)
```

```python
# using or
pd.read_sql("""SELECT * FROM authors WHERE birth_year>=1950 OR country= 'Kenya'

""",conn)
```

```python
# pattern matching
#1. begin with #% ch%
#2. specific letter anywhere of %of%
#3. underscore_(determine the characters an item has)
```

```python
pd.read_sql("""
        SELECT * FROM customers WHERE name LIKE '%z%';
""",conn)
```

```python
#names with 9 characters
pd.read_sql("""
        SELECT * FROM customers WHERE name LIKE '_________';
""",conn)
```

```python
# order by  and limit
pd.read_sql("""
        SELECT * FROM books ORDER BY price DESC, title LIMIT 6;
""",conn)
```

```python
# aggregate function: helps group a number rows into summary numbers
# count(*)(sum), avg, min, max, round
pd.read_sql("""
        SELECT count(*) AS total_books from books;
""",conn)
```

```python
# avg price of books
pd.read_sql("""
        SELECT avg(price) AS avg_price from books;
""",conn)
```

```python
# rounding off ro the nearest decimal place 
pd.read_sql("""
        SELECT ROUND(avg(price),2) AS avg_price from books;
""",conn)
```

```python
# max price of books
pd.read_sql("""
        SELECT max(price) AS max_price from books;
""",conn)
```

```python
# min price of books
pd.read_sql("""
        SELECT min(price) AS min_price from books;
""",conn)
```

```python
# sum price of books
pd.read_sql("""
        SELECT sum(price) AS sum_price from books;
""",conn)
```
