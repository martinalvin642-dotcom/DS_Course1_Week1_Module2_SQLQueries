# DATA ENGINEERING LIFECYCLE

Data scientists often use or start with data that is already available either in csv, format, an API, dashboard etc, but in real production environment, that data doesn't just appear but it goes through a process for it to be available. That process is called **Data Engineering Lifecycle**.<br>

This process is essential to a data scientist since the quality of your data determines the quality of your analysis and the quality of performance of your models.<br>

Data engineering lifecycle has the following connected steps:
- Data Generation
- Data Storage
- Data Ingestion
- Data Transformation
- Data Serving

1. **Data Generation** -> is point where data is captured or generated. Data can be generated using: web scraping, POS, surveys, IOT, mobile apps etc<br>

2. **Data Storage** -> This is saving data that you generated in an organised manner once it has been cleaned.<br>

3. **Data Ingestion** -> Process of loading data or putting data into a system.<br>

4. **Data Transformation** -> Its the process of preparing raw data. It may include: cleaning, removing duplicates, nulls, outliers, filtering, aggregating, merging etc

5. **Data Serving** -> Making your data available. Serving is done through APIs, end-user applications, dashboards, reports

### Data Generation
Simulate sensor data

```python
import numpy as np
import pandas as pd
np.random.seed(23)

sensor_temp_data = np.random.uniform(15,35, (10, 5))
print(sensor_temp_data)
```

### DATA STORAGE

```python
df = [
    {
        "morning": [23,34,21,20,14],
        # "morning": np.random.randint(6)
        "mid_day": [32,23,12,25,17],
        "evening": [12,10,23,19,20]
    }
]
df1 = pd.DataFrame(df)
df1.to_csv("temp_data.csv", index=False)
```

### DATA INGESTION

```python
pd.read_csv("temp_data.csv").head()
```

```python
pd.read_csv("temp_data.csv").shape
```

```python
df1.shape
```

### DATA TRANSFORMATION

```python
df_melted = df1.melt(value_vars=['morning', 'mid_day', 'evening'], \
                    var_name='Time_of_day', value_name='Temperatures')

# unnest the temperatures lis column
df_unnested = df_melted.explode('Temperatures')

# convert temperatures columns to numeric

df_unnested['Temperatures'] = pd.to_numeric(df_unnested['Temperatures'])

df_unnested.head()
```

### DATA SERVING
putting data in a format that is end-user friendly

```python
df1.to_csv("temp_data_cleaned.csv", index=False)
```

DATA PIPELINE

```python
# Step1: Data Generation
 
students = pd.DataFrame({
    "names": ['Steve', 'Reginah', 'Grace', 'Joel'],
    "scores": [56,781,45,np.nan]
})

# Step2: Data storage

students.to_csv("students_info.csv", index=False)

# Step3: Data Ingestion
# Reading into memory
df_students = pd.read_csv("students_info.csv")

# Step4: Data Transformation
df_students['scores'] = df_students['scores'].replace(781, 78)
df_students['scores'].replace(781,78)
df_students['scores'] = df_students['scores'].fillna(df_students['scores'].mean())
df_students
```

### DATA SERVING

```python
df_students.to_csv("clean_student_data.csv", index=False)
```
