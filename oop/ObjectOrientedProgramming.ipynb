## INHERITANCE

Its a principle that allows one class(child) to inherit attributes and methods from another class(parent/super).<br>
This means the superclass inherits properties and behaviors of the super class but can still define its own properties.<br>
example:<br>
> we can have two bank clients, one is a regular client, the other is a premium client, the other is a premium client.<br>
> theres a section that deals with premium clients.<br>
if we were to program that we have 2 classes, one for all clients and the other for premium clients. The latter will inherit from the super.<br>
All clients can have, name,account number etc
a premium will have the above but also have there own properties like stock portfolio.


```python
class BankClient :
    def __init__(self, email, name, account_number, ammount):
        self.name = name
        self.account = account_number
        self.email = email
        self.ammount = ammount
    def withdraw(self,w_ammount):
        remaining_ammount =self.ammount - w_ammount
        print(f"{self.name}, your balance in your account is:{remaining_ammount} .We will invoince you at {self.email} ")

class premiumClient(BankClient):
    def __init__(self, email, name, account_number, stock_potfolio,ammount, stock_value):
        super().__init__(email,name, account_number, ammount)
        self.stock = stock_potfolio
        self.stock_value = stock_value
    def sell(self,s_stock):
        stock_ammmount= s_stock * self.stock_value
        new_amt = stock_ammmount + self.ammount
        total_stock = self.stock - s_stock
        print(f"HI! {self.name}, your new stock count is: {total_stock}, we sent you an invoice at {self.email} you new ammount {new_amt}")
        
alvin = BankClient("martin@.com", "Alvin", 20221, 2000000)
alvin.withdraw(1999999)
moha = premiumClient("Moha@.com", "moha", 294495, 670, 30000000, 19.5)
moha.sell(220)
```
