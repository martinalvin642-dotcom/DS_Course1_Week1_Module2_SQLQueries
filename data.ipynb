Data types: Containers that you choose for your data  to ensure good perfomance, good data integrity and provides sanity for everyone that interacts with your data.

4 types of data types : TEXT, REAL-/FLOAT,  integer and BLOB

1. TEXT : string (characters) e.g email, name, descriptions
2. REAL/FLOAT ()
3. Integer
4. BLOB

```python
# 1. TEXT : string (characters) e.g email, name, descriptions

CREATE TABLE IF NOT EXIST employees(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    descriptions TEXT,
    bio TEXT
    zip_code   :0100  725101365
)

# Any number with leading zeroes at disappears permanetly. so 0100 becomes 100 in your database
 
```

```python
# INTEGER : whole numbers 

CREATE TABLE products(
    id INTERGER PRIMARY KEY,
    name TEXT NOT NULL,
    quantity INTEGER
    times_sold INTEGER
);

```

```python
# REAL : decimal numbers or float numbers in the database

CREATE TABLE trips(
    id INTGER PRIMARY KEY,
    fare REAL,
    distance_km REAL,
    rating REAL
);
```

```python
CREATE TABLE auth(
    user_id INTEGER PRIMARY KEY,
    fingerprint BLOB,
    avatar BLOB,  
)
```
